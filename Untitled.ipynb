```python
import pandas as pd
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_recall_fscore_support
from imblearn.over_sampling import SMOTE, ADASYN,SVMSMOTE,SMOTENC
from sklearn.naive_bayes import GaussianNB , MultinomialNB , ComplementNB ,CategoricalNB
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_classif,mutual_info_classif
from jupyterthemes import jtplot
%matplotlib inline
jtplot.style(theme='onedork', context='notebook', ticks=True, grid=False)
```

```python
np.set_printoptions(suppress=True)
```

```python
#讀去資料
train_data = pd.read_csv('cctx_dataset.csv')
dirty_X = train_data.drop(['alert_key','sar_flag','cust_id','date'],axis = 1)
dirty_Y = train_data.sar_flag
X = dirty_X.drop(['amt','total_asset'],axis = 1)
y = X.describe().loc['max']
```

```python
#資料標準化
sc = MinMaxScaler().fit(dirty_X)
X_dirty_std = sc.transform(dirty_X)
X_dirty_std = pd.DataFrame(X_dirty_std,columns=dirty_X.columns)

X_std = X_dirty_std.drop(['amt','total_asset'],axis = 1)
y_std = X_std.describe().loc['max']



fig,(ax1,ax2) = plt.subplots(1,2,constrained_layout=True,figsize=(12,4))
ax1.barh(X.columns,y)
ax1.set_title('Before Standardization')
ax2.barh(X_std.columns,y_std)
ax2.set_title('After Standardization')
fig.show()
fig.savefig('STD.png')
```

```python
#特徵選取
Select = SelectKBest(f_classif,k=9)
features = Select.fit_transform(X_dirty_std,dirty_Y)
features_names = Select.get_feature_names_out()
df_select = pd.DataFrame([Select.scores_] , columns = X_dirty_std.columns)
df_select.sort_values(by = 0 ,axis=1 ,ascending=True,inplace = True)
bar_colors = ['tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue',
             'tab:blue', 'tab:red', 'tab:red', 'tab:red','tab:red','tab:red','tab:red']
# plt.xlable("sss")
plt.title("ANOVA F-value")
plt.barh(df_select.columns,np.array(df_select.values.reshape(9)),
         color=bar_colors)
# plt.show()
plt.savefig('Select.png')
```

```python
features = np.array(X_dirty_std[features_names])
target = np.array(train_data.sar_flag)
```

```python
#資料分割
X_train_std, X_test_std, y_train, y_test = train_test_split(features,target,test_size=0.3,random_state=0)
y_num_0 = np.sum(y_train == 0)
y_num_1 = np.sum(y_train == 1)
```

```python
# SMOTE
X_train_std_smote , y_train_smote = SVMSMOTE(random_state = 0).fit_resample(X_train_std,y_train)
print(X_train_std.shape)
y_num_smote_0 = np.sum(y_train_smote == 0)
y_num_smote_1 = np.sum(y_train_smote == 1)
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,constrained_layout=True,figsize=(10,4))
ax1.bar(["major","miner"],(y_num_0,y_num_1),width = 0.5)
ax1.set_title('Before SMOTE')
ax2.bar(["major","miner"],(y_num_smote_0,y_num_smote_1),width = 0.5)
ax2.set_title('After SMOTE')
fig.savefig('SMOTE_num.png')
plt.show()
```

```python
#建模訓練
clf = svm.SVC(kernel='rbf' , gamma = 'auto',C=10 , probability=True)
clf.fit(X_train_std,y_train)

clf2 = svm.SVC(kernel='sigmoid' , gamma = 'auto',C=100 ,
               probability=True,class_weight='balanced')
clf2.fit(X_train_std,y_train)

clf3 = svm.SVC(kernel='rbf' , gamma = 'auto',C=10 , probability=True)
clf3.fit(X_train_std_smote,y_train_smote)
```

```python
#預測
pred  = clf.predict(X_test_std)
pred2  = clf2.predict(X_test_std)
pred3  = clf3.predict(X_test_std)
```

```python
#模型評分
aa = clf.score(X_train_std,y_train)
bb = clf.score(X_test_std,y_test)
cc = clf2.score(X_train_std,y_train)
dd = clf2.score(X_test_std,y_test)
ee = clf3.score(X_train_std_smote,y_train_smote)
ff = clf3.score(X_test_std,y_test)
print('svm training score:',aa)
print('svm test score:',bb)
print('svm training score:',cc)
print('svm test score:',dd)
print('svm training score:',ee)
print('svm test score:',ff)
```

```python
s1 = precision_recall_fscore_support(y_test,pred)
s2 = precision_recall_fscore_support(y_test,pred2)
s3 = precision_recall_fscore_support(y_test,pred3)
```

```python
print('precision_svm major:%f  miner:%f '%(s1[0][0],s1[0][1]))
print('recall_svm major:%f  miner:%f '%(s1[1][0],s1[1][1]))
print('precision_decsiontree major:%f  miner:%f '%(s2[0][0],s2[0][1]))
print('recall_decsiontree major:%f  miner:%f '%(s2[1][0],s2[1][1]))
print('precision_decsiontree major:%f  miner:%f '%(s3[0][0],s3[0][1]))
print('recall_decsiontree major:%f  miner:%f '%(s3[1][0],s3[1][1]))
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,constrained_layout=True,figsize=(10,6))
aaa = ax1.bar(["major","minor"],(aa,bb),width = 0.5)
ax1.set_title('orginal')
ax1.set_ylim(0,1.05)
ax1.text("major",aa,s = '%.4f' % aa, ha='center', va= 'bottom',fontsize=15)
ax1.text("minor",bb,s = '%.4f' % bb, ha='center', va= 'bottom',fontsize=15)
ax2.bar(["major","minor"],(cc,dd),width = 0.5)
ax2.set_title('weight balanced')
ax2.set_ylim(0,1.05)
ax2.text("major",cc,s = '%.4f' % cc, ha='center', va= 'bottom',fontsize=15)
ax2.text("minor",dd,s = '%.4f' % dd, ha='center', va= 'bottom',fontsize=15)
ax3.bar(["major","minor"],(ee,ff),width = 0.5)
ax3.set_title('SMOTE')
ax3.set_ylim(0,1.05)
ax3.text("major",ee,s = '%.4f' % ee, ha='center', va= 'bottom',fontsize=15)
ax3.text("minor",ff,s = '%.4f' % ff, ha='center', va= 'bottom',fontsize=15)
fig.savefig('inbalance_accuracy.png')
plt.show()
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,constrained_layout=True,figsize=(10,6))
ax1.bar(["major","minor"],(s1[1][0],s1[1][1]),width = 0.5)
ax1.set_title('orginal')
ax1.set_ylim(0,1.1)
ax1.text("major",s1[1][0],s = '%.4f' % s1[1][0], ha='center', va= 'bottom',fontsize=15)
ax1.text("minor",s1[1][1],s = '%.4f' % s1[1][1], ha='center', va= 'bottom',fontsize=15)
ax2.bar(["major","minor"],(s2[1][0],s2[1][1]),width = 0.5)
ax2.set_title('weight balanced')
ax2.set_ylim(0,1.1)
ax2.text("major",s2[1][0],s = '%.4f' % s2[1][0], ha='center', va= 'bottom',fontsize=15)
ax2.text("minor",s2[1][1],s = '%.4f' % s2[1][1], ha='center', va= 'bottom',fontsize=15)
ax3.bar(["major","minor"],(s3[1][0],s3[1][1]),width = 0.5)
ax3.set_title('SMOTE')
ax3.set_ylim(0,1.1)
ax3.text("major",s3[1][0],s = '%.4f' % s3[1][0], ha='center', va= 'bottom',fontsize=15)
ax3.text("minor",s3[1][1],s = '%.4f' % s3[1][1], ha='center', va= 'bottom',fontsize=15)
fig.savefig('inbalance_recall.png')
plt.show()
```
